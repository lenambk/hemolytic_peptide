```python
import Bio
from Bio.Seq import Seq
from Bio import SeqIO
from Bio.SeqUtils import GC
import pandas as pd
```

TÍNH TOÁN VỚI FILE NEGATIVE

```python
#doc file du lieu, hien thi du lieu va do dai cua chuoi peptit
for seq_record in SeqIO.parse('/content/gdrive/MyDrive/Layer1-negative.fansta', 'fasta'):
  print(seq_record.id)
  print(seq_record.seq, '=>', len(seq_record))
  break
```

```python
# luu cac peptit vao mang peptides
peptides=[]
# luu id vao mang id
id=[]
for seq_record in SeqIO.parse('/content/gdrive/MyDrive/Layer1-negative.fansta', 'fasta'):
  peptides.append(seq_record.seq)
  id.append(seq_record.id)
print(peptides[1])
print(id[1])
```

```python
#tinh chi so AAC
#ham tinh chi so AAC
def AAC(aminoaxit, peptide):
  my_seq=Seq(peptide)
  my_seq.count(aminoaxit)
  return float('{:.4f}'.format(my_seq.count(aminoaxit)/len(my_seq)))
```

```python
#tinh aac voi tung amino axit trong peptide
A=[]
for i in peptides:
  A.append(AAC('A', i))
print(A[2])
R=[]
for i in peptides:
  R.append(AAC('R', i))
N=[]
for i in peptides:
  N.append(AAC('N', i))
D=[]
for i in peptides:
  D.append(AAC('D', i))
C=[]
for i in peptides:
  C.append(AAC('C', i))
E=[]
for i in peptides:
  E.append(AAC('E', i))
Q=[]
for i in peptides:
  Q.append(AAC('Q', i))
G=[]
for i in peptides:
  G.append(AAC('G', i))
H=[]
for i in peptides:
  H.append(AAC('H', i))
I=[]
for i in peptides:
  I.append(AAC('I', i))
L=[]
for i in peptides:
  L.append(AAC('L', i))
K=[]
for i in peptides:
  K.append(AAC('K', i))
M=[]
for i in peptides:
  M.append(AAC('M', i))
F=[]
for i in peptides:
  F.append(AAC('F', i))
P=[]
for i in peptides:
  P.append(AAC('P', i))
S=[]
for i in peptides:
  S.append(AAC('S', i))
T=[]
for i in peptides:
  T.append(AAC('T', i))
W=[]
for i in peptides:
  W.append(AAC('W', i))
Y=[]
for i in peptides:
  Y.append(AAC('Y', i))
V=[]
for i in peptides:
  V.append(AAC('V', i))
```

```python
df0=pd.DataFrame()
df0['id']=id
df0['A']=A
df0['R']=R
df0['N']=N
df0['D']=D
df0['C']=C
df0['E']=E
df0['Q']=Q
df0['G']=G
df0['H']=H
df0['I']=I
df0['L']=L
df0['K']=K
df0['M']=M
df0['F']=F
df0['P']=P
df0['S']=S
df0['T']=T
df0['W']=W
df0['Y']=Y
df0['V']=V
df0['label']=0
```

```python
print(df0)
df0.to_csv('AAC_0.csv')
```
