```python
import Bio
from Bio.Seq import Seq
from Bio import SeqIO
from Bio.SeqUtils import GC
import pandas as pd
```

TÍNH TOÁN VỚI FILE NEGATIVE 

```python
#doc file du lieu, hien thi du lieu va do dai cua chuoi peptit
for seq_record in SeqIO.parse('/content/gdrive/MyDrive/Layer1-negative.fansta', 'fasta'):
  print(seq_record.id)
  print(seq_record.seq, '=>', len(seq_record))
  break
```

```python
# luu cac peptit vao mang peptides
peptides=[]
# luu id vao mang id
id=[]
for seq_record in SeqIO.parse('/content/gdrive/MyDrive/Layer1-negative.fansta', 'fasta'):
  peptides.append(seq_record.seq)
  id.append(seq_record.id)
print(peptides[1])
print(id[1])
```

```python
#tinh chi so DPC
#ham tinh chi so DPC
def DPC(diaminoaxit, peptide):
  my_seq=Seq(peptide)
  my_seq.count(diaminoaxit)
  return float('{:.4f}'.format(my_seq.count(diaminoaxit)/(len(my_seq)-1)))
```

```python
def DP(**kw):
  for key, value in kw.items():
    return ("%s%s" %(key, value))
aminoaxit=['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']
name_col=[]
for i in aminoaxit:
  name_col.append(DP(A=i))
for i in aminoaxit:
  name_col.append(DP(C=i))
for i in aminoaxit:
  name_col.append(DP(D=i))
for i in aminoaxit:
  name_col.append(DP(E=i))
for i in aminoaxit:
  name_col.append(DP(F=i))
for i in aminoaxit:
  name_col.append(DP(G=i))
for i in aminoaxit:
  name_col.append(DP(H=i))
for i in aminoaxit:
  name_col.append(DP(I=i))
for i in aminoaxit:
  name_col.append(DP(K=i))
for i in aminoaxit:
  name_col.append(DP(L=i))
for i in aminoaxit:
  name_col.append(DP(M=i))
for i in aminoaxit:
  name_col.append(DP(N=i))
for i in aminoaxit:
  name_col.append(DP(P=i))
for i in aminoaxit:
  name_col.append(DP(Q=i))
for i in aminoaxit:
  name_col.append(DP(R=i))
for i in aminoaxit:
  name_col.append(DP(S=i))
for i in aminoaxit:
  name_col.append(DP(T=i))
for i in aminoaxit:
  name_col.append(DP(V=i))
for i in aminoaxit:
  name_col.append(DP(W=i))
for i in aminoaxit:
  name_col.append(DP(Y=i))
print(name_col)
```

```python
diaminoaxit = name_col
df=pd.DataFrame()
df['id']=id
for i in diaminoaxit:
  for j in name_col:
    j=[]
    for k in peptides:
      j.append(DPC(i, k))
    df[i]=j
df
```

```python
df['label']=0
df.to_csv('DPC_0.csv')
```
