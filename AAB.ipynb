```python
seq="/content/peptides0.txt"
result_filename= "pfeature_result.csv"

aab(seq,'tempfile_out')
df = pd.read_csv('tempfile_out')
df1 = df.iloc[:,:-1].fillna('N')
df1.to_csv(result_filename,index=None) 
df2 = df.dropna(axis="columns", how="any")
df2['label']=0
df2.to_csv('AAB_0.csv',index=None)
os.remove('tempfile_out')
os.remove('pfeature_result.csv')
```
