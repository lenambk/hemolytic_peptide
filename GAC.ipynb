```python
import Bio
from Bio.Seq import Seq
from Bio import SeqIO
from Bio.SeqUtils import GC
import pandas as pd
```

```python
# NHÓM BÉO
G1=['I', 'M', 'L', 'V', 'A', 'G']
# NHÓM THƠM
G2=['W', 'Y', 'F']
# NHÓM ĐIỆN TÍCH DƯƠNG
G3=['H', 'R', 'K']
# NHÓM ĐIỆN TÍCH ÂM
G4=['E', 'D']
# NHÓM KHÔNG TÍCH ĐIỆN
G5=['Q', 'N', 'P', 'C', 'T', 'S'] 
```

TÍNH TOÁN VỚI FILE NEGATIVENEGATIVE

```python
#doc file du lieu, hien thi du lieu va do dai cua chuoi peptit
for seq_record in SeqIO.parse('/content/gdrive/MyDrive/Layer1-negative.fansta', 'fasta'):
  print(seq_record.id)
  print(seq_record.seq, '=>', len(seq_record))
  break
```

```python
# luu cac peptit vao mang peptides
peptides=[]
# luu id vao mang id
id=[]
for seq_record in SeqIO.parse('/content/gdrive/MyDrive/Layer1-negative.fansta', 'fasta'):
  peptides.append(seq_record.seq)
  id.append(seq_record.id)
print(peptides[3])
print(id[3])
```

```python
#tinh thanh phan tung nhom theo 5 chi so
#ham tinh thanh phan 
def GAC(group, peptide):
  my_seq=Seq(peptide)
  sum=0
  for i in group:
    sum=sum+my_seq.count(i)
  s=float('{:.4f}'.format(sum/len(my_seq)))
  return (s)
```

```python
group=[G1, G2, G3, G4, G5]
header=['GAC1', 'GAC2', 'GAC3', 'GAC4', 'GAC5']
```

```python
df_0 = pd.DataFrame()
df_0['id'] = id
header_1=header
for i in range(len(group)):
  for j in header:
    j=[]
    for k in peptides:
      j.append(GAC(group[i], k))
    df_0[header_1[i]]=j
df_0['label']=0
df_0
```

```python
df_0.to_csv('GAC_0.csv')
```
