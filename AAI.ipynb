```python
import Bio
from Bio.Seq import Seq
from Bio import SeqIO
from Bio.SeqUtils import GC
import pandas as pd
```

TÍNH TOÁN VỚI FILE NEGATIVE


```python
#doc file du lieu, hien thi du lieu va do dai cua chuoi peptit
for seq_record in SeqIO.parse('/content/gdrive/MyDrive/Layer1-negative.fansta', 'fasta'):
  print(seq_record.id)
  print(seq_record.seq, '=>', len(seq_record))
  break
```

```python
# luu cac peptit vao mang peptides
peptides=[]
# luu id vao mang id
id=[]
for seq_record in SeqIO.parse('/content/gdrive/MyDrive/Layer1-negative.fansta', 'fasta'):
  peptides.append(seq_record.seq)
  id.append(seq_record.id)
print(peptides[1])
print(id[1])
```

```python
#hiển thị cơ sở dữ liệu axit amin index
file_index = pd.read_csv('/content/gdrive/MyDrive/aaindex.csv')
file_index
```

```python
import csv
with open('/content/gdrive/MyDrive/aaindex.csv') as f:
    reader = csv.reader(f)
    l= [row for row in reader]
print(l[1][1])
```

```python
header = []
for i in range(len(file_index['INDEX'])):
  header.append(l[i][0])
print(header)
```

```python
AA = 'ARNDCQEGHILKMFPSTWYV'
index={}
for i in range(1, (len(AA)+1)):
  index[i]=AA[i-1]
print(index)
```

```python
#hàm tính toán chỉ số aai ứng với từng chỉ số và peptit tương ứng
def AAI(peptide, k):
  sum = 0
  for i in range(1, 21):
    for j in range(len(peptide)):
      if peptide[j]==index[i]:
        sum = sum+float(l[k][i])
        aai = float('{:4f}'.format(sum/len(peptide)))
  return aai
```

```python
name_df = []
for i in range(len(file_index['INDEX'])):
  name_df.append(l[i][0])

df = pd.DataFrame()
df['id'] = id
for i in range(1, len(header)):
  header[i] = []
  for j in range(len(peptides)):
    header[i].append(AAI(peptides[j], i))
  df[name_df[i]] = header[i]
df
```

```python
df['label'] = '0'
```

```python
df.to_csv('AAI_0.csv')
```
