COMPOSITION

```python
#tinh thanh phan tung nhom theo 7 chi so
#ham tinh thanh phan 
def composition(group, peptide):
  my_seq=Seq(peptide)
  sum=0
  for i in group:
    sum=sum+my_seq.count(i)
  s=float('{:.4f}'.format(sum/len(my_seq)))
  return (s)
```

```python
aa = ['1', '2', '3']
bb = ['HB', 'NV', 'PO', 'PZ', 'CH', 'SS', 'SA']
group=[GHB1, GNV1, GPO1, GPZ1, GCH1, GSS1,  GSA1, GHB2, GNV2, GPO2, GPZ2, GCH2, GSS2,  GSA2, GHB3, GNV3, GPO3, GPZ3, GCH3, GSS3,  GSA3]
header_c = []
for a in aa:
  for b in bb:
    header_c.append('CTD_'+ b + a)
print(header_c)
```

```python
df_c0 = pd.DataFrame()
df_c0['id'] = id
header_c1=header_c
for i in range(len(group)):
  for j in header_c:
    j=[]
    for k in peptides:
      j.append(composition(group[i], k))
    df_c0[header_c1[i]]=j
df_c0['label']=0
df_c0
```

```python
#hàm tính distribution
def distribution(p, CTD_):
  CTD_D=[]
  for i in CTD_:
    j=float('{:.4f}'.format(i*p))
    CTD_D.append(j)
  return CTD_D
```

```python
#hàm tính distribution
def distribution(p, CTD_):
  CTD_D=[]
  for i in CTD_:
    j=float('{:.4f}'.format(i*p))
    CTD_D.append(j)
  return CTD_D
```

```python
header_d0 = []
header_d25=[]
header_d50=[]
header_d75=[]
header_d100=[]

bb=['HB1', 'HB2', 'HB3', 'NV1', 'NV2', 'NV3', 'PO1', 'PO2', 'PO3', 'PZ1', 'PZ2', 'PZ3', 'CH1', 'CH2', 'CH3', 'SS1', 'SS2', 'SS3', 'SA1', 'SA2', 'SA3']
for b in bb:
	header_d0.append('CTD_' + '0' + '_'+ b)
for b in bb:
	header_d25.append('CTD_' + '25' + '_'+ b)
for b in bb:
	header_d50.append('CTD_' + '50' + '_'+ b)
for b in bb:
	header_d75.append('CTD_' + '75' + '_'+ b)
for b in bb:
	header_d100.append('CTD_' + '100' + '_'+ b)
print(header_d0)
```

```python
#tính số aminoaxit của mỗi nhóm có trong peptit
def composition(group, peptide):
  my_seq=Seq(peptide)
  sum=0
  for i in group:
    sum=sum+my_seq.count(i)
  s=sum
  return s
aa = ['1', '2', '3']
bb = ['HB', 'NV', 'PO', 'PZ', 'CH', 'SS', 'SA']
group=[GHB1, GNV1, GPO1, GPZ1, GCH1, GSS1,  GSA1, GHB2, GNV2, GPO2, GPZ2, GCH2, GSS2,  GSA2, GHB3, GNV3, GPO3, GPZ3, GCH3, GSS3,  GSA3]
header_c = []
for a in aa:
  for b in bb:
    header_c.append('CTD_'+ b + a)
#print(header_c)
df_c1 = pd.DataFrame()
df_c1['id'] = id
header_c1=header_c
for i in range(len(group)):
  for j in header_c:
    j=[]
    for k in peptides:
      j.append(composition(group[i], k))
    df_c1[header_c1[i]]=j
df_c1
```
